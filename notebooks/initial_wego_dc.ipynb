```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
with open('../data/WeGo.json') as fi:
    credentials = json.load(fi)
```

```python
Authorization = credentials['Authorization']
```

```python
URL = 'https://api.goswift.ly/speed-map/nashville/route/23'

params = {
    'direction': '1',
    'startDate': '05-16-2022',
    'endDate'  : '06-15-2022',
    'resolution': 'hiRes',
}

headers = {'Authorization': Authorization,
          'Content-Type': 'application/json'}
```

```python
response = requests.get(URL, params = params, headers = headers)
```

```python
response
```

```python
res = response.json()
res
```

```python
inbound_res_df = pd.DataFrame(res['data']['segments'])
inbound_res_df.sort_values('pathLengthMeters')
```

```python
inbound_res_df.sort_values('countForSegment')
```
