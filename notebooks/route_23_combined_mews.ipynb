```python
combined_map = folium.Map(location =  [36.2, -86.7775], zoom_start = 12)


route_details_map = folium.FeatureGroup(name='Route 23')

stops_in = folium.FeatureGroup(name='Inbound Bus Stops', overlay=True, show=False)

stops_out = folium.FeatureGroup(name='Outbound Bus Stops', overlay=True, show=False)

am_in = folium.FeatureGroup(name='Morning Rush Inbound', overlay=True, show=False)

pm_ou = folium.FeatureGroup(name='Afternoon Rush Outbound', overlay=True, show=False)

em_in = folium.FeatureGroup(name='Early Morning Inbound', overlay=True, show=False)

em_ou = folium.FeatureGroup(name='Early Morning Outbound', overlay=True, show=False)

opd_in = folium.FeatureGroup(name='Daytime Offpeak Inbound', overlay=True, show=False)

opd_ou = folium.FeatureGroup(name='Daytime Offpeak Outbound', overlay=True, show=False)

ln_in = folium.FeatureGroup(name='Evening/Late Night Inbound', overlay=True, show=False)

ln_ou = folium.FeatureGroup(name='Evening/Late Night Outbound', overlay=True, show=False)



route_details_map.add_to(combined_map)

stops_in.add_to(combined_map)

stops_out.add_to(combined_map)

am_in.add_to(combined_map)

pm_ou.add_to(combined_map)

em_in.add_to(combined_map)

em_ou.add_to(combined_map)

opd_in.add_to(combined_map)

opd_ou.add_to(combined_map)

ln_in.add_to(combined_map)

ln_ou.add_to(combined_map)


folium.LayerControl(position='topleft', collapsed=False).add_to(combined_map)

style1 = {'color': 'gray', 'weight' : 2} #overall route
style2 = {'color': 'yellow', 'weight' : 7} #outbound during school months
style3 = {'color': 'red', 'weight' : 2} #outbound during summer months
style4 = {'color': 'blue', 'weight' : 7} #inbound during school months
style5 = {'color': 'orange', 'weight' : 2} #inbound during summer months


folium.GeoJson(route_details_json, style_function=lambda x: style1).add_to(route_details_map)

folium.GeoJson(slow_am_scho_geojson, style_function=lambda x: style4).add_to(am_in)
folium.GeoJson(slow_am_sum_geojson, style_function=lambda x: style5).add_to(am_in)

folium.GeoJson(slow_pmr_inschool.set_crs("EPSG:4326"), style_function=lambda x: style2).add_to(pm_ou)
folium.GeoJson(slow_pmr_summer.set_crs("EPSG:4326"), style_function=lambda x: style3).add_to(pm_ou)

folium.GeoJson(slow_emi_scho_geojson, style_function=lambda x: style4).add_to(em_in)
folium.GeoJson(slow_emi_sum_geojson, style_function=lambda x: style5).add_to(em_in)

folium.GeoJson(slow_emo_scho_geojson, style_function=lambda x: style2).add_to(em_ou)
folium.GeoJson(slow_emo_sum_geojson, style_function=lambda x: style3).add_to(em_ou)

folium.GeoJson(slow_offpin_inschool.set_crs("EPSG:4326"), style_function=lambda x: style4).add_to(opd_in)
folium.GeoJson(slow_offpin_inschool.set_crs("EPSG:4326"), style_function=lambda x: style5).add_to(opd_in)

folium.GeoJson(slow_do_in_school.set_crs("EPSG:4326"), style_function=lambda x: style2).add_to(opd_ou)
folium.GeoJson(slow_do_summer.set_crs("EPSG:4326"), style_function=lambda x: style3).add_to(opd_ou)

folium.GeoJson(slow_lni_scho_geojson, style_function=lambda x: style4).add_to(ln_in)
folium.GeoJson(slow_lni_sum_geojson, style_function=lambda x: style5).add_to(ln_in)

folium.GeoJson(slow_lno_inschool.set_crs("EPSG:4326"), style_function=lambda x: style2).add_to(ln_ou)
folium.GeoJson(slow_lno_summer.set_crs("EPSG:4326"), style_function=lambda x: style3).add_to(ln_ou)

for row_index, row_values in inbound_stop_details.iterrows():
    loc = [row_values['geometry'].y, row_values['geometry'].x]
    popi = str(row_values['stopName'])
    iframe = folium.IFrame(popi)
    pop = folium.Popup(iframe, min_width=150, max_width=200)
    icon= folium.Icon(color="gray",icon="bus", prefix='fa')
    
    marker = folium.Marker(
        location = loc, 
        popup = pop, 
        icon = icon) 

    marker.add_to(stops_in)

for row_index, row_values in outbound_stops.iterrows():
    loc = [row_values['geometry'].y, row_values['geometry'].x]
    popi = str(row_values['stopName'])
    iframe = folium.IFrame(popi)
    pop = folium.Popup(iframe, min_width=150, max_width=200)
    icon= folium.Icon(color="gray",icon="bus", prefix='fa')
    
    marker = folium.Marker(
        location = loc, 
        popup = pop, 
        icon = icon) 

    marker.add_to(stops_out)
    
combined_map
```
